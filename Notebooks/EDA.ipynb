```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
# Assigned values to region
dataset.loc[:122,'Region'] = 0
dataset.loc[122:,'Region'] = 1
```

```python
dataset.head()
```

```python
df = dataset
```

```python
df[['Region']]=df[['Region']].astype('int')
```

```python
df.info()
```

```python
# removed null values
df = df.dropna().reset_index(drop = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(index=122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#fix spaces in col name
df.columns = df.columns.str.strip()
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
#changing other cols to float
[features for features in df.columns if df[features].dtypes == 'O']
```

```python
df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']] = df[['Rain', 'FFMC', 'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype('float')
```

```python
df.info()
```

```python
df.to_csv('Cleaned Dataset',index=False)
```
